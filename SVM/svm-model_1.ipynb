ORDINAL ENCODER

```python
ode_cols = ['LotShape', 'LandContour','Utilities','LandSlope',  'BsmtQual',  'BsmtFinType1',  'CentralAir',  'Functional', 'GarageFinish', 'GarageQual', 'PavedDrive', 'ExterCond', 'KitchenQual', 'BsmtExposure', 'HeatingQC','ExterQual', 'BsmtCond']
```

ONE HOT ENCODER

```python
ohe_cols = ['Street', 'LotConfig','Neighborhood', 'Condition1', 'Condition2', 'BldgType', 'HouseStyle', 'RoofStyle', 'Exterior1st', 'Exterior2nd','Foundation',  'Electrical',  'SaleType', 'MSZoning', 'SaleCondition', 'Heating', 'GarageType', 'RoofMatl']
```

```python
num_pipeline = Pipeline(steps=[
    ('impute', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])
```

```python
ode_pipeline = Pipeline(steps=[
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('ode', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1))
])
```

```python
ohe_pipeline = Pipeline(steps=[
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('ohe', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])
```

```python
col_trans = ColumnTransformer(transformers=[
    ('num_p', num_pipeline, num_cols),
    ('ode_p', ode_pipeline, ode_cols),
    ('ohe_p', ohe_pipeline, ohe_cols),
    ],
    remainder='passthrough', 
    n_jobs=-1)
```

```python
pipeline = Pipeline(steps=[
    ('preprocessing', col_trans)
])
```

```python
X = np.linspace(0, 10, 100).reshape(-1, 1)
y = np.sin(X).ravel()

# Fit SVR model
svr = SVR(kernel='rbf')
svr.fit(X, y)

# Make predictions
y_pred = svr.predict(X)

# Plot data and predictions
plt.scatter(X, y, color=gradient_colors[0], label='Data')
plt.plot(X, y_pred, color=gradient_colors[29], label='SVR Prediction')
plt.xlabel('X')
plt.ylabel('y')
plt.title('SVR Prediction')
plt.legend()
plt.show()
```

```python
def preprocess_data():
    # Get file path from user input
    file_path = filedialog.askopenfilename(title="Select CSV File")

    # Load data using the file path
    data = pd.read_csv(file_path)

    # Check data types before OneHotEncoder
    for col in data.select_dtypes(include=[object]):
        try:
            # Attempt conversion to numeric (might raise errors for invalid strings)
            pd.to_numeric(data[col])
        except (ValueError, TypeError):  # Handle potential conversion errors
            print(f"Warning: Column '{col}' contains strings that might not be convertible to numbers. Consider data cleaning or excluding this column from OneHotEncoder.")
            # Handle non-numeric values here

    # Define categorical and numeric columns
    categorical_features = ['MSZoning', 'Street', 'LotShape', 'LandContour', 'Utilities',
                            'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2',
                            'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st',
                            'Exterior2nd', 'ExterQual', 'ExterCond', 'Foundation', 'BsmtQual',
                            'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'Heating', 'HeatingQC',
                            'CentralAir', 'Electrical', 'KitchenQual', 'Functional', 'GarageType',
                            'GarageFinish', 'GarageQual', 'PavedDrive', 'SaleType',
                            'SaleCondition']
    
    # Handle missing values in categorical columns
    data[categorical_features] = data[categorical_features].fillna('Unknown')
    
    numeric_features = data.select_dtypes(include=['int64', 'float64']).columns

    ode_cols = ['LotShape', 'LandContour','Utilities','LandSlope',  'BsmtQual',  'BsmtFinType1',  'CentralAir',  'Functional', 'GarageFinish', 'GarageQual', 'PavedDrive', 'ExterCond', 'KitchenQual', 'BsmtExposure', 'HeatingQC','ExterQual', 'BsmtCond']
    ohe_cols = ['Street', 'LotConfig','Neighborhood', 'Condition1', 'Condition2', 'BldgType', 'HouseStyle', 'RoofStyle', 'Exterior1st', 'Exterior2nd','Foundation',  'Electrical',  'SaleType', 'MSZoning', 'SaleCondition', 'Heating', 'GarageType', 'RoofMatl']
    num_pipeline = Pipeline(steps=[
    ('impute', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
    ])
    ode_pipeline = Pipeline(steps=[
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('ode', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1))
    ])
    ohe_pipeline = Pipeline(steps=[
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('ohe', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
    ])
    col_transformer = ColumnTransformer(transformers=[
    ('num_p', num_pipeline, num_cols),
    ('ode_p', ode_pipeline, ode_cols),
    ('ohe_p', ohe_pipeline, ohe_cols),
    ],
    remainder='passthrough',
    n_jobs=-1)

    # Handle non-numeric values in numeric columns
    data[numeric_features] = data[numeric_features].apply(pd.to_numeric, errors='coerce')

    # Fit the ColumnTransformer to the loaded data
    col_transformer.fit(data)

    # Transform the data using the fitted col_transformer
    preprocessed_data = col_transformer.transform(data.copy())

    # Display success message
    success_label.config(text="Preprocessing Successful!")

    # Return the file path of the encoded CSV file
    return file_path
```
